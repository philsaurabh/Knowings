```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.tsa.stattools import adfuller
```

```python
# 1. Load the dataset which is indexed by the month.
import pandas as pd
dataset=pd.read_csv("AirPassengers.csv", index_col="Month")
dataset.head()
```

```python
# 2. Review the data statistics
dataset.describe()
```

```python
# 3. Check for datatype in dataset
dataset.dtypes
```

```python
# 4. Convert Month object into datetime
df=pd.read_csv("AirPassengers.csv")
df['Month'] = pd.to_datetime(df['Month'])
df.tail()
```

```python
df = df.set_index("Month")
df.head()
```

```python
# 5. Cleaning and preparing time series data
df.info()
```

```python
#No null values. So converting na values to zero
```

```python
df.nunique()
```

```python
# 6.  Check for Stationarity of time series
```

```python
#Split the data
df_1=df[0:73]
df_2=df[73:144]
```

```python
#Mean of data
df_1.mean()
```

```python
df_2.mean()
```

```python
#Variance of data
df_1.var()
```

```python
df_2.var()
```

```python
#ADF statistic to check stationarity
result=[0,0]
t = df["#Passengers"].values
timeseries = adfuller(t)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in timeseries[4].items():
 print('\t%s: %.3f' % (key, value))
if timeseries[0] > timeseries[4]["5%"]:
  print ("Failed to Reject Ho - Time Series is Non-Stationary")
else:
 print ("Reject Ho - Time Series is Stationary")
```
